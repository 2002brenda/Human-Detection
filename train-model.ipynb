```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import os
from ultralytics import YOLO

# Path model YOLOv8 yang sudah dilatih
model_path = r"runs\detect\train\weights\best.pt"

# Path gambar asli yang ingin diprediksi
image_path = "D:\People-Detection-YOLOv8n-main\orang5.png"

# Load model YOLOv8
model = YOLO(model_path)

# Jalankan prediksi dan simpan hasilnya
results = model.predict(source=image_path, save=True, imgsz=250)

# Path gambar hasil prediksi (YOLO menyimpan hasil di runs/detect/predict/)
predicted_image_path = r"runs\detect\predict4\orang5.jpg"

# Cek apakah file hasil prediksi sudah dibuat
if not os.path.exists(predicted_image_path):
    print("File hasil prediksi belum tersedia. Pastikan YOLO berhasil menyimpan gambar.")
else:
    # Load gambar hasil prediksi langsung dari output YOLOv8
    predicted_image = cv2.imread(predicted_image_path)
    predicted_image = cv2.cvtColor(predicted_image, cv2.COLOR_BGR2RGB)

    # Tampilkan hasil prediksi langsung di output notebook
    plt.figure(figsize=(8, 6))  # Atur ukuran gambar agar terlihat jelas
    plt.imshow(predicted_image)
    plt.axis("off")  # Hilangkan sumbu koordinat
    plt.title("Hasil Prediksi YOLOv8")
    plt.show()
```
